```python
import requests
import time
import pandas as pd

HOST = 'http://localhost:8080'
```

```python
# Ex 2.2.A
response = requests.get(HOST + '/devices')
if response.status_code == 200:
    # print(response.json())
    response_a = response.json()
    devices = response_a.get('mac_addresses', [])
    print(devices)
else:
    print('The server is offline.')
```

```python
# Ex 2.2.B
payload = {
    'from' : 0,
    'to' : round(time.time() * 1000) #Time
}
response_df = pd.DataFrame(columns=['mac_address','timestamps','battery_levels'])
for device in devices:
    response_b = requests.get(HOST + '/device/' + device , params=payload)
    response_json = response_b.json()
    mac_address = device
    timestamps = response_json.get('timestamps', [])
    battery_levels = response_json.get('battery_levels', [])
    if len(timestamps) > 0 and len(battery_levels) == len(timestamps):
        response_df = response_df.append(pd.DataFrame({
            'mac_addresses' : mac_address,
            'timestamps' : timestamps,
            'battery_levels' : battery_levels

        }))
response_df['timestamps'] = pd.to_datetime(response_df['timestamps'], unit='ms')
```

```python
# 2.2.C
response_c = requests.delete(HOST+'/device/'+devices[-1])
if response_c.status_code == 200:
    print(f'Successfully deleted last device data {devices[-1]}')
```
